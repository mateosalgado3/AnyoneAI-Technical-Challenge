**Question 1:** How many customers are in the file?

```python
from os.path import exists
import csv

# check if file can be found and open the file
datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

count = 0

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    # skip header
    next(csvreader)

    # count rows
    for row in csvreader:
        if row:  # evita filas vacías
            count += 1

print("Total customers:", count)
```

**Question 2:** In how many different states do the customers live in?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

states = set()

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)

    for row in csvreader:
        if row:
            state = row[4].strip().upper()  #NORMALIZATION
            if state:
                states.add(state)

print("Different states:", len(states))
print("States:", states)
```

**Question 3** What is the state with most customers?



```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

state_count = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            state = row[4].strip().upper()

            if state:
                if state in state_count:
                    state_count[state] += 1
                else:
                    state_count[state] = 1

# find state
max_state = None
max_count = 0

for state, count in state_count.items():
    if count > max_count:
        max_count = count
        max_state = state

print("State with most customers:", max_state)
print("Number of customers:", max_count)
```

**Question 4** What is the state with the least customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

state_count = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            state = row[4].strip().upper()

            if state:
                state_count[state] = state_count.get(state, 0) + 1

# find state with minimum customers
min_state = min(state_count, key=state_count.get)
min_count = state_count[min_state]

print("State with least customers:", min_state)
print("Number of customers:", min_count)
```

**Question 5:** What is the most common last name?




```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

last_name_count = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            # normalize last name
            last_name = row[2].strip().upper()

            if last_name:
                last_name_count[last_name] = last_name_count.get(last_name, 0) + 1

# get most common last name
most_common_last = max(last_name_count, key=last_name_count.get)
count = last_name_count[most_common_last]

print("Most common last name:", most_common_last)
print("Count:", count)
```

**Question 1:** How many unique orders are in the orders.csv file?

```python
# open and evaluate orders.csv

from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

order_ids = set()

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            order_id = row[1].strip()
            if order_id:
                order_ids.add(order_id)

print("Unique orders:", len(order_ids))
```

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

```python
from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

order_items = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            order_id = row[1].strip()

            if order_id:
                order_items[order_id] = order_items.get(order_id, 0) + 1

# compute average
total_items = sum(order_items.values())
total_orders = len(order_items)

average = total_items / total_orders

print("Average items per order:", round(average, 2))
```

```python
# find max items in a single order
max_items = max(order_items.values())

print("Highest number of items in a single order:", max_items)
```

**Question #4:** What is the number of orders placed in October 2021?


```python
from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

orders_oct_2021 = set()

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    next(csvreader)  # skip header

    for row in csvreader:
        if row:
            order_id = row[1].strip()
            date = row[2].strip()

            # filter October 2021
            if date.startswith("2021-10"):
                orders_oct_2021.add(order_id)

print("Orders in October 2021:", len(orders_oct_2021))
```

**Question #5:** Which customer spent the most amount of money in 2021?


```python
from os.path import exists
import csv

orders_file = "./sample_data/orders.csv"
customers_file = "./sample_data/customers.csv"

if not exists(orders_file) or not exists(customers_file):
    raise SystemExit("Run first cell to download data")

# Load customer names
customers = {}

with open(customers_file, "r") as fl:
    csvreader = csv.reader(fl, delimiter=",")
    next(csvreader)  # skip header

    for row in csvreader:
        if row and len(row) >= 3:
            customer_id = row[0].strip()
            first_name = row[1].strip()
            last_name = row[2].strip()

            customers[customer_id] = f"{first_name} {last_name}"

# Compute spending in 2021
customer_spending = {}
seen_orders = set()

with open(orders_file, "r") as fl:
    csvreader = csv.reader(fl, delimiter=",")
    next(csvreader)  # skip header

    for row in csvreader:
        if row and len(row) >= 4:
            customer_id = row[0].strip()
            order_id = row[1].strip()
            date = row[2].strip()
            order_total = row[3].strip()

            if date.startswith("2021"):
                # Count each order only once
                if order_id and order_id not in seen_orders:
                    seen_orders.add(order_id)

                    if customer_id and order_total:
                        total = float(order_total)
                        customer_spending[customer_id] = customer_spending.get(customer_id, 0) + total

# Get top customer
top_customer_id = max(customer_spending, key=customer_spending.get)
top_amount = customer_spending[top_customer_id]
top_customer_name = customers.get(top_customer_id, "Unknown")

print("Top customer ID:", top_customer_id)
print("Top customer name:", top_customer_name)
print("Total spent:", top_amount)
```

**Question #6:** Historically, what is the best month for sales?

```python
from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("Run first cell to download data")

monthly_sales = {}
seen_orders = set()

month_names = {
    "01": "January",
    "02": "February",
    "03": "March",
    "04": "April",
    "05": "May",
    "06": "June",
    "07": "July",
    "08": "August",
    "09": "September",
    "10": "October",
    "11": "November",
    "12": "December"
}

with open(datafile, "r") as fl:
    csvreader = csv.reader(fl, delimiter=",")

    next(csvreader)  # skip header

    for row in csvreader:
        if row and len(row) >= 4:
            order_id = row[1].strip()
            date = row[2].strip()
            order_total = row[3].strip()

            # count each order only once
            if order_id and order_id not in seen_orders:
                seen_orders.add(order_id)

                # validate date and total
                if len(date) >= 7 and order_total:
                    month_number = date[5:7]

                    if month_number in month_names:
                        total = float(order_total)
                        month_name = month_names[month_number]
                        monthly_sales[month_name] = monthly_sales.get(month_name, 0) + total

# find best historical month
best_month = max(monthly_sales, key=monthly_sales.get)
best_sales = monthly_sales[best_month]

print("Best month for sales:", best_month)
print("Total sales:", best_sales)
print(monthly_sales)
```
